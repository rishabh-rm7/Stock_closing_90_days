## part- 1 from yahoo finance, get_data_yahoo (attribute)

```python
import pandas as pd
```

```python
import pandas_datareader as pdr
```

```python
colname = ['Stockname', 'SYMBOL']
excl_data = pd.read_excel('stocks.xlsx', sheet_name = 'Sheet2', names = colname)
excl_data1 = excl_data.copy()
excl_data1.head()
```

```python
df = []
df = pd.DataFrame(df)

df['Symbols'] = excl_data1['SYMBOL']
```

```python
symbol_list = list(df['Symbols'])
symbol_list
```

```python
new_symbol_list = []
for i in symbol_list:
    res = " ".join(i.split())
    new_symbol_list.append(res)
```

```python
new_symbol_list
```

```python
import datetime
import sys
from pandas_datareader._utils import RemoteDataError
```

```python
data2 = []
data2 = pd.DataFrame(data2)
for symbol in new_symbol_list:
    try:
        sym = pdr.get_data_yahoo(symbol,
                                start=datetime.datetime(2021,8,14),
                                end=datetime.datetime(2021,10,14))
        data2[symbol + '_close'] = sym['Close']
    except (RemoteDataError,KeyError ) :
        continue
```

# part - 2 from nsepy library and get_history attribute

```python
from nsepy import get_history
from datetime import date
```

```python
data3 = []
data3 = pd.DataFrame(data3)
for symbol in new_symbol_list:
    sym = get_history(symbol, start = date(2021,8,14), end = date(2021,10,14))
    data3[symbol + '_close'] = sym['Close']
    
```
